```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset=pd.read_csv('student_scores.csv')
```

```python
#shows the first 5 rows in the data
dataset.head()
```

```python
#we describe the data inorder to undertstand the tabelled data 

dataset.describe()
```

```python
#plot a 2-D graph for knowing the relationships between hours and scores
dataset.plot(x='Hours', y='Scores', style='o')
plt.title('Hours vs Percentage')
plt.xlabel('Hours Studied')
plt.ylabel('Percentage Score')
plt.show()
```

```python
# as in a positive linear regression y=mx+c where m-slope and c-intercept
#attribute-hours
# label-predicted score
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 1].values
```

```python
#train_test_split() method
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# training the algorithm
#y=mx+c
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# to find the best fit
#retriving the intercept-c
print(regressor.intercept_)
```

```python
#retriving the slope-m
print(regressor.coef_)
```

```python
#making predictions
y_pred = regressor.predict(X_test)

#compare the actual output values as y_pred is a numphy array
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df
```

```python
# as we see the predicted values and the actual values though are not tat accurate.
```

```python
#evaluating the algorithm
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
#the root mean squared error is 4.64,
#which is less than 10% of the mean value of the percentages of all the students i.e. 51.48. 
#algorithm is proper.
```
